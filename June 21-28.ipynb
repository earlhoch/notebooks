```python
import numpy
import math
import string
import matplotlib.pyplot as plt
%matplotlib inline

def logit(x):
    return math.log(x/(1-x))

bothTest = []
bothTrain = []
TrainTest = []
TestTrain = []

logitBothTest = []
logitBothTrain = []
logitTrainTest = []
logitTestTrain = []

full = []
logitFull = []

fileIn = open('scored_models_sorted.dat', 'r')

for line in fileIn:
    
    split = string.split(line)
    
    name = str(split[2])
    name = string.split(name, '.')[0]
    
    model0Value = float(split[6])
    model1Value = float(split[7])
    
    inModelXTest = split[9]
    inModelYTest = split[10]
    
    pair = (model0Value, model1Value, name)
    logitPair = (logit(model0Value), logit(model1Value), name)
    
    
    
    if inModelXTest == '0' and inModelYTest == '0':
        
        bothTrain.append(pair)
        logitBothTrain.append(logitPair)
        
    elif inModelXTest == '1' and inModelYTest == '1':
        
        bothTest.append(pair)
        logitBothTest.append(logitPair)
        
    elif inModelXTest == '0' and inModelYTest == '1':
        
        TrainTest.append(pair)
        logitTrainTest.append(logitPair)
        
    elif inModelXTest == '1' and inModelYTest == '0':
        
        TestTrain.append(pair)
        logitTestTrain.append(logitPair)
    
    full.append(pair)
    logitFull.append(logitPair)

plt.figure(figsize=(15,15))

x = []
y = []

for pair in logitBothTrain:
    x.append(pair[0])
    y.append(pair[1])
plt.scatter(x,y, color = "black", s=50, label = "Both Train", marker = "s", alpha = 0.75)

x = []
y = []

for pair in logitBothTest:
    x.append(pair[0])
    y.append(pair[1])
plt.scatter(x,y, color = "red", s=50, alpha = 0.5, marker = "^")

x = []
y = []

for pair in logitTestTrain:
    x.append(pair[0])
    y.append(pair[1])
plt.scatter(x,y, color = "blue", s=50, label = "Model 0 Test, Model 1 Train", alpha = 0.75)

x = []
y = []

for pair in logitTrainTest:
    x.append(pair[0])
    y.append(pair[1])
plt.scatter(x,y, color = "red", s=50, label = 'Model 0 Train, Model 1 Test', alpha = 0.75)
 
x = []
y = []


x = [-20,20]
y = [-20,20]
plt.plot(x,y, '--', alpha = 0.75, label = "y=x")

plt.legend(fontsize = 20, loc = 4)
plt.xlabel('Model 0', fontsize = 30)
plt.ylabel('Model 1', fontsize = 30)
plt.xlim(-15, 10)
plt.ylim(-15, 10)

x = []
y = []

for pair in logitFull:
    x.append(pair[0])
    y.append(pair[1])

corr = ""
corr = numpy.corrcoef(x = x, y = y)[0][1]
corr = "r = %5f" % corr

plt.text(1, -10.75, corr, fontsize = 30)

#for pair in logitTestTrain:

    #plt.annotate(pair[2], (pair[0],pair[1]))
    
plt.show()
```
